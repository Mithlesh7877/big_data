```
# File location and type
file_location = "/FileStore/tables/1000_Sales_Records.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "false"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema",'true') \
  .option("header", 'true') \
  .option("sep", delimiter) \
  .load(file_location)

# change header by using header with true
```

```
# Create a view or table

temp_table_name = "1000_Sales_Records_csv"

df.createOrReplaceTempView(temp_table_name)
```

```
from pyspark.sql import SparkSession

spark = SparkSession \
    .builder \
    .appName("Basic").getOrCreate()
```

```
# %sql

# /* Query the created temp table in a SQL cell */

# select * from `1000_Sales_Records_csv`
df = spark.sql("select * from 1000_Sales_Records_csv")
```
